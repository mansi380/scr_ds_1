```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\admin\Desktop\New folder\World Population Growth.csv')

print(df.head())

```

```python
if 'gender' in df.columns:
    plt.figure(figsize=(8, 6))
    df['gender'].value_counts().plot(kind='bar', color='skyblue')
    plt.title('Gender Distribution in Population')
    plt.xlabel('Gender')
    plt.ylabel('Frequency')
    plt.xticks(rotation=0)  # To keep the x-axis labels horizontal
    plt.show()
```

```python
if 'age' in df.columns:
    plt.figure(figsize=(8, 6))
    df['age'].dropna().plot(kind='hist', bins=20, color='orange', edgecolor='black')  # Drop NaN values
    plt.title('Age Distribution in Population')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plotting the population distribution by year
df = pd.read_csv(r'C:\Users\admin\Desktop\New folder\World Population Growth.csv')
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Population'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Distribution Over the Years')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting a Histogram for Yearly Growth %
plt.figure(figsize=(10, 6))
plt.hist(df['Yearly Growth %'], bins=20, color='orange', edgecolor='black')
plt.xlabel('Yearly Growth %')
plt.ylabel('Frequency')
plt.title('Distribution of Yearly Population Growth %')
plt.tight_layout()
plt.show()
```
